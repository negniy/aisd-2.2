## Выводы

Метод сортировки выбором более стабильный и всреднем всегда дает одинаковое количество сравнений и копирований
Метод шейкерной сортировки хорошо подходит для почти отсортированных массивов, т.к. тогда происходит минимальное копирование, но очень медленно работает с плохо-сортированными массивами большой длинны. А метод сортировки расческой является самым быстрым, так как при большом кол-ве сортируемых данных выполняет наименьшее кол-во операций по сравнению с двумя другими.
